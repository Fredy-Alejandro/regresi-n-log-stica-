```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
data = pd.read_csv('heart_cleveland_upload.csv')

# Resumen estadístico
print(data.describe())

# Visualización de valores atípicos
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.show()

# Función para encontrar valores atípicos
def find_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
    return outliers
    

# Identificar valores atípicos
outliers = find_outliers(data)
print(outliers)

# Eliminar valores atípicos
data_cleaned = data[~data.isin(outliers)].dropna()
print(data_cleaned.describe())

```

```python
from sklearn.feature_selection import SelectKBest, chi2

X = df.iloc[:, :-1]  # Todas las columnas menos la última (variable objetivo)
y = df.iloc[:, -1]   # Última columna como la variable objetivo

selector = SelectKBest(chi2, k=10)  # Seleccionamos las 10 características más relevantes
X_new = selector.fit_transform(X, y)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
data = pd.read_csv('heart_cleveland_upload.csv')

# Resumen estadístico
print(data.describe())

# Visualización de valores atípicos
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.show()

# Función para encontrar valores atípicos
def find_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
    return outliers
    

# Identificar valores atípicos
outliers = find_outliers(data)
print(outliers)

# Eliminar valores atípicos
data_cleaned = data[~data.isin(outliers)].dropna()
print(data_cleaned.describe())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = pd.read_csv('heart_cleveland_upload.csv')

# Resumen inicial de los datos
print(df.describe())
print(df.info())

# Definir variables independientes (X) y la variable dependiente (y)
X = df.drop('condition', axis=1)
y = df['condition']

# Visualización de relaciones entre variables
sns.pairplot(df)
plt.show()

# visualizacion de relaciones entre variables con pairplot
sns.pairplot(df,diag_kind='kde',markers='+')
plt.show()


            
```

```python
# Limpiar datos y tratar valores faltantes
df.dropna(inplace=True)

# Codificación de variables categóricas si es necesario
df = pd.get_dummies(df, drop_first=True)

# Mostrar el dataset preprocesado
print(df.head())
```

```python
# Identificar valores faltantes
print(df.isnull().sum())

# Rellenar valores faltantes (ejemplo: con la media)
df = df.fillna(df.mean())

# Transformación de variables (ejemplo: estandarización)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

X = df.iloc[:, :-1]  # Todas las columnas menos la última (variable objetivo)
y = df.iloc[:, -1]   # Última columna como la variable objetivo

selector = SelectKBest(chi2, k=10)  # Seleccionamos las 10 características más relevantes
X_new = selector.fit_transform(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
```

```python
# Selección de características más relevantes
from sklearn.feature_selection import SelectKBest, chi2
selector = SelectKBest(score_func=chi2, k=10)  # Selecciona las 10 características más relevantes
X_new = selector.fit_transform(X, y)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.show()
```
